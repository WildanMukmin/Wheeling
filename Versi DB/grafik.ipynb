```python
import matplotlib.pyplot as plt
from collections import defaultdict
from database import Database

def plot_assessments(db):
    data = db.get_assessments_data()
    
    if not data:
        print("No assessment data found.")
        return
    
    students_data = defaultdict(lambda: defaultdict(list))
    
    for name, date, aspect, score in data:
        students_data[name][aspect].append((date, score))
    
    for student, aspects in students_data.items():
        plt.figure(figsize=(10, 6))
        plt.title(f"Assessment Scores for {student}")
        plt.xlabel("Date")
        plt.ylabel("Score")
        
        for aspect, scores in aspects.items():
            dates, scores = zip(*scores)
            plt.plot(dates, scores, marker='o', label=aspect.capitalize())
        
        plt.legend()
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    db = Database()
    plot_assessments(db)
```
